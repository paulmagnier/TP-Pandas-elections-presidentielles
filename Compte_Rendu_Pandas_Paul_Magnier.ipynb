```python
from pandas import *
import matplotlib.pyplot as plt

data_presidentielles = pandas.read_csv("elections-presidentielles-2017-1ertour.csv",  sep =';')
#print(data_presidentielles)

# Question 1

jlm = data_presidentielles["MÉLENCHON Jean-Luc"]
total = jlm.sum()
print("Nombre total d'électeurs de Mélenchon : ", total)
```

```python
par_arrond = data_presidentielles[["NUM_ARROND", "MÉLENCHON Jean-Luc"]].groupby("NUM_ARROND").sum()
par_arrond
```

```python
s = [i for i in range(1, 21)]
data = par_arrond.reindex(pandas.Index(s))
data
```

```python
donnees = data_presidentielles[["NUM_ARROND", "LE PEN Marine", "MACRON Emmanuel", "MÉLENCHON Jean-Luc", "FILLON François"]].groupby("NUM_ARROND").sum()
n = 2# ici on choisit l'arrondissement souhaité
donnees = donnees.loc[n]
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.pie(donnees, labels = ["LE PEN", "MACRON", "MELENCHON", "FILLON"])
ax2 = donnees.plot.bar()
plt.show()
```

```python
data = data_presidentielles[["NUM_ARROND", "NB_BL", "NB_NUL", "NB.EXPRIM"]].groupby("NUM_ARROND").mean()
def affiche(arr):
    data2 = data.loc[arr]
    print(data2)
    data2.plot.bar()
affiche(2)
```
